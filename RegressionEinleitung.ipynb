```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 16, 10
```

```python
x= np.linspace(0,2*np.pi,1000)
```

```python
y= 4*x+ 10 + np.random.ranf(1000)
y2 = np.sin(x) + np.random.ranf(1000)
```

```python
plt.plot(x,y,'ro')
plt.plot(x,y2,'go')
```
